```python
def cost_custom(d):
    cost = None
    
    if -4 <= d <= 4:

        cost = 0.0

    if d > 4:
        
        cost = abs(d - 4) ** 2

    if d < -4:

        cost = abs(4 + d) ** 2
       
    
    return cost


```

```python
cost_fit = np.zeros((10000))
numbers = np.linspace(-16,16,10000)
for i in range(10000):
    
    cost_fit[i] = cost_custom(numbers[i])


# numbers = np.sort(numbers)
# print(numbers)
plt.scatter(numbers,cost_fit)
```

```python
x = MX.sym('x')
y = MX.sym('y')
vx = MX.sym('vx')
vy = MX.sym('vy')
v = MX.sym('v')
t_inv = MX.sym('t_inv')
yaw = MX.sym('yaw')

dt = MX.sym('dt')
l = MX.sym('l')
p = MX.sym('p')
cd = MX.sym('cd')
cf = MX.sym('cf')
cc = MX.sym('cc')

acc = MX.sym('acc')
delta = MX.sym('delta')

vel = Function('vel', [vx, vy], [sqrt(vx**2 + vy**2)])

tan_inv = Function('tan_inv', [vx, vy], [atan2(vy, vx)])

pred_x = Function('pred_x', [x, yaw, vx, vy, v, delta, dt], [x + v * cos(atan2(tan(delta), 2) + yaw) * dt])

pred_y = Function('pred_y', [y, yaw, vx, vy, v, delta, dt], [y + v * sin(atan2(tan(delta), 2) + yaw) * dt])

pred_yaw = Function('pred_yaw', [yaw, vx, vy, v, delta, l, dt], [yaw + (v * tan(delta) * dt / sqrt((l**2) + ((0.5 * l * tan(delta))**2)))])

pred_vx = Function('pred_vx', [vx, vy, v, acc, p, cd, cf, dt], [vx + ((p * acc) - (cd * v * vx - cf * vx)) * dt])

pred_vy = Function('pred_vy', [vx, vy, v, t_inv, delta, cd, cf, cc, dt], [vy - (cc * (t_inv - delta) + (cd * v + cf) * vy) * dt])

opti = Opti()
a = opti.variable()
d = opti.variable()

opti.minimize((pred_x(0, 0, 0, 0, vel(0, 0), d, 0.3) - 0) ** 2 + (pred_y(0, 0, 0, 0, vel(0, 0), d, 0.3) - 0) ** 2 + (pred_yaw(0, 0, 0, 0, d, 4.45, 0.3) - 0) ** 2 + (pred_vx(0, 0, 0, a, 4.22, -0.0013, 0.362, 0.03) - 2.0) ** 2 + (pred_vy(0, 0, vel(0, 0), tan_inv(0, 0), d, -0.0013, 0.362, 1.318, 0.3) - 0) ** 2)

opti.subject_to(opti.bounded(0, a, 1))

opti.solver('ipopt')
sol = opti.solve()
print(sol.value(a))
```
